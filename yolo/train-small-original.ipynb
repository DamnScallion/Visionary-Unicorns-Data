```python
import matplotlib.pyplot as plt
from ultralytics import YOLO
import os
```

```python
ROOT_DIR = '/content/Visionary-Unicorns-Project/yolo/config'
```

# Trainning YOLO with our dataset

```python
# Create a new YOLO model from scratch
# Below is YOLOv8 Small size model architecture
model = YOLO('yolov8s.pt')
```

```python
# Train the small size yolo model with original given dataset for 100 epochs
results = model.train(data=os.path.join(ROOT_DIR, 'yolo_config.yaml'), epochs=100, imgsz=224, name='yolov8s_default_original')
```
